* **Competencia Desventajosa**: En un mercado globalizado y competitivo, otros importadores y negocios ya establecidos cuentan con identidades corporativas sólidas y plataformas digitales robustas. La falta de estas herramientas coloca a este nuevo emprendimiento en una **clara desventaja**, limitando su capacidad de escalar y captar nuevos clientes.

* **Pérdida de Oportunidades de Posicionamiento SEO**: Sin una página web, el negocio no puede beneficiarse del posicionamiento en buscadores (**SEO**), lo que significa que es invisible para clientes que buscan productos o servicios específicos a través de Google u otros motores de búsqueda.

# Métodos de Promtping Utilizados

En el desarrollo del proyecto, se emplearon diferentes enfoques de prompting con el fin de optimizar la generación de resultados esperados. A continuación, se detallan los dos métodos principales utilizados:

**Prompting de una sola solicitud (One-shot prompting):**
Este enfoque consiste en proporcionar una única entrada o solicitud para obtener una respuesta o conjunto de respuestas. La técnica se basa en la presentación de ejemplos mínimos o una descripción precisa, lo que permite generar resultados basados en la información dada. El objetivo principal es lograr una respuesta adecuada sin requerir múltiples iteraciones, garantizando eficiencia en el proceso de solicitud.

**Prompting iterativo:**
Se utiliza en situaciones donde se requiere un refinamiento continuo del resultado a través de la repetición o ajustes sucesivos del prompt. Este método permite realizar ajustes incrementales en el contexto de la solicitud, con el fin de ir adaptando la respuesta a medida que se avanza en el proceso de generación. A través de esta técnica, se busca una mayor precisión y especificidad en los resultados finales.

**Prompting dirigido:**
Dentro del diccionario de datos, se aplica este enfoque para crear eslóganes o generar contenidos específicos. El prompting dirigido implica una formulación precisa de la solicitud, de manera que se pueda obtener el resultado deseado de forma clara y eficiente. En este caso, se ajustan las entradas de manera que se guíe al modelo hacia la generación de resultados que sigan un formato o línea específica, lo que incrementa la probabilidad de obtener el resultado esperado con alta precisión.

```python
# Importamos las dependencias de OpenAI y demas
import openai
import requests
import os
```

```python
# Agregamos nuestra clave de API de OpenAI
openai.api_key = "" 
```

```python
# Creamos el contexto de nuestro entorno de solicitud
context = """
SL Importados es una empresa emergente con un perfil multifacético que se dedica a la importación de productos en los siguientes sectores clave:

Ferretería

Electrodomésticos

Energías renovables (especialmente paneles solares)

Soluciones térmicas móviles (como calderas diésel para camiones y casas rodantes)

Movilidad urbana (bicicletas)

La empresa busca consolidarse como un referente confiable, innovador y sostenible, tanto en el mercado nacional como internacional. Su enfoque se basa en la diversificación inteligente, la calidad de sus productos, la adaptabilidad a las tendencias del mercado y un compromiso firme con el desarrollo responsable y sustentable.

Actualmente, SL Importaciones carece de una identidad corporativa digital definida, lo que representa una oportunidad estratégica para construir una imagen sólida y coherente que:

Potencie su presencia en canales digitales.

Transmita profesionalismo, modernidad y confianza.

Refleje la diversidad y el dinamismo de su oferta.

Genere diferenciación frente a la competencia.

Atraiga clientes, aliados estratégicos y distribuidores.

Tu misión como especialista en branding, identidad corporativa y diseño digital con enfoque estratégico es desarrollar una propuesta integral que articule los elementos centrales de la marca en su versión digital, con base en los valores, el propósito y la proyección futura de SL Importaciones.
"""
```

```python
# Diccionario de prompts donde la clave es el nombre del prompt y el valor es el texto (Esto permite tener una forma más robusta y legible de interactuar con tus diferentes instrucciones e iterar.)
prompts_con_nombre = {
    "MisionVision" : """Redactá una misión y una visión para SL Importados que reflejen una identidad corporativa sólida, profesional y con proyección estratégica a largo plazo. La misión debe expresar claramente el propósito actual de la empresa, su propuesta de valor y su compromiso con los clientes. La visión debe proyectar una imagen aspiracional de futuro, alineada con el crecimiento sostenido, la innovación y el liderazgo en el sector de importación de productos ferreteros, energías renovables y soluciones de movilidad.""",
"Valores": """Identificá y describí cinco valores fundamentales que definan la cultura organizacional de SL Importaciones. Estos valores deben reflejar su compromiso con la calidad, la innovación, la sostenibilidad y la excelencia en el servicio, y estar alineados con su propósito estratégico de posicionarse como un actor confiable y visionario en el mercado nacio""",
"Eslogan" : """ "Importamos innovación. Potenciamos tu futuro." Este tiene que ser el slogan """,   
"Tipografia": """Teniendo en cuenta que SL Importaciones actualmente utiliza un logotipo con las iniciales "SL" en tipografía sólida y moderna, acompañado por una línea curva en tonos negro y naranja que transmite dinamismo y proyección, sugerí una paleta de colores corporativos y una tipografía institucional que refuercen esta identidad visual. La propuesta debe ser coherente con una imagen empresarial profesional, versátil y orientada al crecimiento futuro. Los colores y tipografías sugeridos deben transmitir confianza, innovación, solidez y adaptabilidad, alineados con los rubros que abarca la empresa (ferretería, energías renovables, movilidad y soluciones térmicas).""",     
"Redes" :  """Teniendo en cuenta que SL Importaciones utiliza un logotipo con las iniciales "SL" en tipografía robusta y moderna, acompañado por una línea curva en tonos negro y naranja que transmite dinamismo y solidez, proponé una paleta de colores corporativos complementaria y una tipografía institucional coherente con esa identidad visual. Además, detallá ejemplos concretos de cómo aplicar estos elementos en distintos soportes digitales y comunicacionales (como sitio web, redes sociales, presentaciones corporativas, firmas de correo, catálogos o packaging), de modo que refuercen la imagen profesional, confiable y moderna de la empresa. La propuesta debe alinearse con una visión empresarial de largo plazo, adaptable a diversos mercados y rubros como ferretería, electrodomesticos, energías renovables y movilidad."""   
    
}
```

```python
# Itera a través de los prompts y sus nombres (Utilizando el diccionario creado anteriormente)
for nombre_prompt, texto_prompt in prompts_con_nombre.items():
    print(f"--- Procesando prompt: {nombre_prompt} ---") # Muestra el nombre del prompt
    conversation = [
        {"role": "system", "content": context},
        {"role": "user", "content": texto_prompt} #'texto_prompt'
    ]

    response = openai.ChatCompletion.create(
        model='gpt-3.5-turbo',
        messages=conversation,
        max_tokens=1000
    )

    message = response.choices[0]['message']
    print(f"Rol: {message['role']}")
    print(f"Contenido: {message['content']}")
    print("-" * 30) # Separador para mayor claridad
```

```python
!pip install google.genai
from google import genai
from google.genai import types
from PIL import Image
from io import BytesIO
import base64

```

```python
#Generacion de logo
client = genai.Client(api_key='')

image_text= """Crea un logotipo moderno y versátil para una empresa de importaciones llamada "SL Importados". El logotipo debe inspirarse en la imagen proporcionada y mantener la esencia de las letras "SL" entrelazadas o visualmente conectadas.

Elementos clave a conservar y refinar:

Letras "SL": Deben ser el elemento central y más destacado del logotipo. Explora opciones para hacerlas más limpias, geométricas o estilizadas, manteniendo su legibilidad. Considera un diseño donde las letras se complementen o se integren de forma creativa.
Elemento curvo/circular naranja: Este elemento debe ser reinterpretado de una manera más moderna y adaptable. Podría ser un trazo limpio, una forma geométrica abstracta que rodee o interactúe con las letras "SL", o incluso fragmentos de un círculo que aporten dinamismo. Asegúrate de que el color naranja (o una tonalidad similar) se mantenga como un color secundario distintivo.
Consideraciones de diseño:

Estilo general: El logotipo debe transmitir profesionalismo, dinamismo y modernidad, reflejando la actividad de importación y la variedad de sectores que abarca la empresa (ferretería, electrodomésticos, energías renovables, etc.).
Colores: Mantén el negro como color principal para las letras "SL" y utiliza el naranja como color secundario para el elemento gráfico. Considera la posibilidad de incluir un color neutro adicional (como un gris o blanco) para fondos o detalles, debajo de las letras debes incluir la palabra "Importados"
Tipografía (implícita): Aunque el prompt se centra en el diseño del logo, piensa en una tipografía sans-serif moderna y legible que podría complementar el logotipo en otros elementos de la marca.
Versatilidad: El diseño debe ser adaptable a diferentes tamaños y formatos digitales e impresos (página web, redes sociales, tarjetas de presentación, etc.). Considera la creación de una versión horizontal y vertical si es apropiado.
Impacto visual: El logotipo debe ser memorable y fácilmente reconocible.
Resultado esperado:

Un diseño de logotipo digital, limpio y moderno que conserve la identidad visual básica de la imagen proporcionada, enfatizando las letras "SL" y el elemento naranja debe estar ubicado por detras de las siglas, pero con una estética más profesional y adaptable para una marca de importaciones diversificada.
"""

contents = (image_text)

response = client.models.generate_content(
    model="gemini-2.0-flash-preview-image-generation",
    contents=contents,
    config=types.GenerateContentConfig(
      response_modalities=['TEXT', 'IMAGE']
    )
)

for part in response.candidates[0].content.parts:
  if part.text is not None:
    print(part.text)
  elif part.inline_data is not None:
    image = Image.open(BytesIO((part.inline_data.data)))
    image.save('gemini-native-image.png')
    image.show()
```
